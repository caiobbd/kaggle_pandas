```python
fruity_occurrance = reviews.description.map(lambda x: 'fruity' in x).sum()
tropical_occurrance = reviews.description.map(lambda x: 'tropical' in x).sum()

#disciptor_counts = fruity_occurrance.value_counts() + tropical_occurrance.value_counts()
discriptor_counts1 = pd.Series([tropical_occurrance,fruity_occurrance ], index=['tropical','fruity'])
discriptor_counts1
```

```python

n_trop = reviews.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts2 = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])


discriptor_counts1 == descriptor_counts2


```

```python
def starsystem(row):
    if row.title == 'Canadian Vintners Association':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1



star_ratings = reviews.apply(starsystem, axis = 'columns')


```
